```python
import os

# Caminho para a pasta 'results'
results_path = 'results'

# Itera sobre cada pasta dentro de 'results'
for folder in os.listdir(results_path):
    folder_path = os.path.join(results_path, folder)
    
    # Verifica se é uma pasta
    if os.path.isdir(folder_path):
        npy_count = 0
        
        # Caminha recursivamente pelas subpastas
        for root, dirs, files in os.walk(folder_path):
            # Conta os arquivos .npy na pasta atual
            npy_count += len([f for f in files if f.endswith('.npy')])
        
        print(f'Pasta: {folder}, Arquivos .npy: {npy_count}')
```
