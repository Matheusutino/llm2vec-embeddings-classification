```python
import pandas as pd
from src.core.utils import get_all_files_in_directory

# Obter todos os arquivos no diretório datasets
datasets_path = get_all_files_in_directory("datasets")

# Criar uma lista para armazenar os resultados de cada dataset
results_list = []

# Iterar por cada dataset e calcular as métricas desejadas
for dataset_path in datasets_path:
    df = pd.read_csv(dataset_path)
    
    # Calcular métricas
    num_instances = len(df)
    lengths = df['text'].str.len()
    min_length = lengths.min()
    average_length = lengths.mean()
    max_length = lengths.max()
    variance_length = lengths.std()  # Calcular a variância dos comprimentos de texto
    unique_labels = df['class'].unique()
    num_labels = len(unique_labels)
    
    # Adicionar as métricas ao results_list
    results_list.append({
        'Dataset': dataset_path,
        'Num Instances': num_instances,
        'Min Length': min_length,
        'Average Length': average_length,
        'Max Length': max_length,
        'Variance Length': variance_length,
        'Unique Labels': unique_labels,
        'Num Labels': num_labels
    })

# Criar um DataFrame a partir da lista de resultados
results = pd.DataFrame(results_list)

# Exibir o DataFrame final
results
```

```python
import pandas as pd
from src.core.utils import get_all_files_in_directory

# Obter todos os arquivos no diretório datasets
datasets_path = get_all_files_in_directory("datasets")

# Criar uma lista para armazenar os resultados de cada dataset
results_list = []

# Iterar por cada dataset e calcular as métricas desejadas
for dataset_path in datasets_path:
    dataset_name = dataset_path.split('/')[1][:-4]
    df = pd.read_csv(dataset_path)
    
    # Calcular métricas
    num_instances = len(df)
    lengths = df['text'].str.len()
    min_length = lengths.min()
    max_length = lengths.max()
    median_length = lengths.median()
    variance_length = lengths.std()  # Calcular a variância dos comprimentos de texto
    unique_labels = df['class'].unique()
    num_labels = len(unique_labels)
    
    # Adicionar as métricas ao results_list
    results_list.append({
        'Dataset': dataset_name[:1].upper() + dataset_name[1:],
        'Num Instances': num_instances,
        'Min Length': min_length,
        'Max Length': max_length,
        'Median Length': int(median_length),
        #'Variance Length': variance_length,
        #'Unique Labels': unique_labels,
        'Num Labels': num_labels
    })

# Criar um DataFrame a partir da lista de resultados
results = pd.DataFrame(results_list)
results = results.sort_values(by="Num Instances")

results.to_csv("results/dataset_infos.csv", index=False)
# Exibir o DataFrame final
results
```
